# E-Commerce Behavioral Analytics & Recommendation Engine

## Project Goal

This project builds an end-to-end Big Data pipeline for e-commerce behavioral analytics.

The system processes raw web log events such as:

- `view`
- `cart`
- `purchase`

Using Apache Spark, the system performs:

1. Market Basket Analysis
2. User Affinity Aggregation
3. NoSQL storage in MongoDB
4. Cart Abandonment Recovery
5. Recommendation / campaign decision demo

## Dataset Description

The raw dataset contains more than 10 million e-commerce events.

Important columns:

| Column | Meaning |
|---|---|
| timestamp | Time of user event |
| session_id | User session identifier |
| user_id | Customer identifier |
| event_type | view, cart, or purchase |
| product_id | Product/item identifier |
| price | Product price |
| user_metadata | JSON data about user device/tier/location |
| product_metadata | JSON data about product category/brand/stock |

Because the dataset is large, Pandas is avoided. Spark is used for distributed processing.
